```python
from openai import OpenAI

api_key = '...'
client = OpenAI(api_key=api_key)
```

## 1차 데이터 generation

```python
def call_model(text):
    completion = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "넌 약에 대해 잘 알고 있는 약사야."},
            {"role": "user", "content": f'{text}'}
        ]
    )
    answer = completion.choices[0].message.content
    return answer
```

```python
txt_data_all_1 = []
tmp = []
# for data in data_list:
    # print(data)
data = data_list[0]
file = data_dir + data
# print(file)
f = open(file, 'r')     # mode = 부분은 생략해도 됨
lines = f.readlines()
# print(lines)

for line in lines:
    tmp.append(line)
    if line == '\n':
        t_data = "".join(tmp)
        txt_data_all_1.append(t_data)
        tmp = []
    
    # txt_data
```

```python
import random
```

```python
random.seed(9)
```

```python
txt_data_1 = random.sample(txt_data_all_1, 100)
txt_data_1
```

```python
answer_data_1 = []
for data in tqdm(txt_data_1):
    answer = call_model(data)
    answer_data_1.append(answer)
    
```
