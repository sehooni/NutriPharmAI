```python
import faiss
import pandas as pd
```

```python
df = pd.read_csv("../dataset/nutrient_data_dropped.tsv", sep="\t")
```

```python
df
```

```python
name_embedding = []
for name in df["이름"]:
  emb = client.embeddings.create(
    model = "solar-embedding-1-large-query",
    input = name).data[0].embedding
  name_embedding.append(emb)
name_embedding = np.array(name_embedding)
```

```python

quantizer = faiss.IndexFlatIP(4096)
kor_index = faiss.IndexIVFFlat(quantizer, 4096,
                              10,
                              faiss.METRIC_INNER_PRODUCT)
kor_index.train(name_embedding)
kor_index.add(name_embedding)
```

```python
text = "코큐텐"
query_emb = client.embeddings.create(
    model = "solar-embedding-1-large-query",
    input = text).data[0].embedding
query_emb = np.array(query_emb)

query_emb = np.array([query_emb])
result = kor_index.search(query_emb, 10)[1]

kor_index.search(query_emb,10)
```

```python
from predibase import Predibase, FinetuningConfig, DeploymentConfig

pb = Predibase(api_token="...")
```

```python
dataset = pb.datasets.from_file("./nutrient.jsonl", name="nutrient_data")
```

```python
repo = pb.repos.create(name="nutrient_model", description="nutrient finetuned model", exists_ok=True)

# Start a fine-tuning job, blocks until training is finished
adapter = pb.adapters.create(
    config=FinetuningConfig(
        base_model="solar-1-mini-chat-240612"
    ),
    dataset=dataset, # Also accepts the dataset name as a string
    repo=repo,
    description="initial model with defaults"
)
```
